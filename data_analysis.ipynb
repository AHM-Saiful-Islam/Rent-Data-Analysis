```python
## Basic Modeling
# Train a model: linear regression
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# dataset
X = data[['Size', 'BHK']]
y = data['Rent'] # target

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(f'Intercept: {model.intercept_}')
print(f'Coefficients: {model.coef_}')
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Calculate Mean Squared Error (MSE)
# MSE is a measure of how far off your predictions are from the actual values. 
# The larger the MSE, the more inaccurate your model is in predicting the target variable.
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error (MSE): {mse}')
```

```python
# Root Mean Squared Error (RMSE)
rmse = mse ** 0.5
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

```python
# R-squared (R): Indicates how well the model explains the variance in the target variable.
r_squared = model.score(X_test, y_test)
print(f'R-squared: {r_squared}')
```

```python
## Observations
# the MSE is 2,395,041,835.33, which is a relatively large value. 
# This suggests that the predictions from the model have considerable error.
# An R-squared value of 0.319 means that 31.9% of the variability in your target 
# variable can be explained by the features in your model. This is a relatively 
# low R value, suggesting that the model is not doing a great job at explaining 
# the variation in the target variable. In other words, 68.1% of the variation 
# in the target variable is unexplained by the model.

## The low R-squared value and high MSE indicate that the current model might 
## not be very effective. 
```

```python
## KNN Regression Model
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

# dataset
X = data[['Size', 'BHK']]
y = data['Rent'] # target

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize KNN Regressor with k=5 neighbors
model = KNeighborsRegressor(n_neighbors=5)
model.fit(X_train, y_train)
```

```python
# prediction
y_pred = model.predict(X_test)
```

```python
# Evaluate the Model: We can evaluate the model using Mean Squared Error (MSE) and 
# R-squared (R).

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared: {r2}')
```

```python
# The high **Mean Squared Error (MSE)** indicates that the KNN model's predictions 
# are quite far from the actual values on average, suggesting poor predictive performance.
# The negative **R-squared** value implies that the model performs worse than a simple 
# mean prediction, indicating that it is not fitting the data well and may require 
# adjustments or a different algorithm.
```

```python
## Random Forest Regressor
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# dataset
X = data[['Size', 'BHK']]
y = data['Rent'] # target

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared: {r2}')
```

```python
# The high MSE and negative R-squared indicate that the Random Forest model is not 
# performing well, suggesting that the features or data preprocessing may need to be 
# revisited.
```

```python
# Gradient Boosting Regressor
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# dataset
X = data[['Size', 'BHK']]
y = data['Rent'] # target

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Gradient Boosting Regressor
model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared: {r2}')
```

```python
# The Gradient Boosting model shows a moderate improvement with a lower MSE and 
# a positive R-squared, but it still indicates underperformance and may need 
# further tuning or feature engineering.
```
